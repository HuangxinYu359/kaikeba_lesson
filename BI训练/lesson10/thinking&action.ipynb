```python
import numpy as np
a = np.array(
[
    [0,0,0,1/3,0,0],
    [1/3,0,0,0,1/2,0],
    [0,1,0,1/3,1/2,0],
    [1/3,0,0,0,0,1],
    [0,0,1,1/3,0,0],
    [1/3,0,0,0,0,0]
])
b = np.array([1/6,1/6,1/6,1/6,1/6,1/6])
w = b
def work(a,w):
    for i in range(100):
        w = np.dot(a,w)
        print(w)
def random_work(a,w,n):
    d = 0.85
    for i in range(100):
        w = (1-d)/n+d*np.dot(a,w)
        print(w)
```

```python
work(a,w)
```

```python
random_work(a,w,n=4)
```

```python
import jieba
import jieba.analyse
import jieba.posseg as pseg
```

```python
with open('G:/project_courseware/核心班BI/lesson10/L10/textrank/news_高考.txt',encoding ='utf-8' ) as file:
    data = file.read()
```

```python
data
```

```python
seg_list = jieba.cut(data,cut_all=False)
print(' '.join(seg_list))
```

```python
words = pseg.cut(data)
for word,flag in words:
    print('%s, %s' %(word ,flag))
```

```python
# 通过TF-IDF获取关键词
keywords = jieba.analyse.extract_tags(data, topK=20, withWeight=True, allowPOS=('n','nr','ns'))
for item in keywords:
    print(item[0],item[1])
print('-'*100)
```
