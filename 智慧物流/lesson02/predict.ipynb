```python
train_data['fraud']
x_fraud = train_data.loc[:,train_data.columns != 'fraud']
y_fraud = train_data['fraud']
x_fraud
```

```python
x_late = train_data.loc[:,train_data.columns !='late_delivery']
y_late = train_data['late_delivery']
```

```python
x_o = train_data.loc[:,train_data.columns !='Order Item Quantity']
y_o = train_data['Order Item Quantity']
```

```python
#数据切分
from sklearn.model_selection import train_test_split
x_f_train,x_f_test,y_f_train,y_f_test =train_test_split(x_fraud,y_fraud,test_size = 0.2)
x_l_train,x_l_test,y_l_train,y_l_test =train_test_split(x_late,y_late,test_size = 0.2)
x_s_train,x_s_test,y_s_train,y_s_test =train_test_split(x_s,y_s,test_size = 0.2)
x_o_train,x_o_test,y_o_train,y_o_test =train_test_split(x_o,y_o,test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
#数据归一化
sc = StandardScaler()
x_f_train = sc.fit_transform(x_f_train)
x_f_test = sc.transform(x_f_test)
x_l_train = sc.fit_transform(x_l_train)
x_l_test = sc.transform(x_l_test)
x_s_train = sc.fit_transform(x_s_train)
x_s_test = sc.transform(x_s_test)
x_o_train = sc.fit_transform(x_o_train)
x_o_test = sc.transform(x_o_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,f1_score
def model_stats(model,x_train,x_test,y_train,y_test,name = 'Fraud'):
    model = model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_pred,y_test)
    recall = recall_score(y_pred,y_test)
    auc=roc_auc_score(y_pred,y_test)
    f1 = f1_score(y_pred,y_test)
    print('Model Used:',model)
    print('{} Accuracy :{}%'.format(name,accuracy *100))
    print('{} Recall :{}%'.format(name,recall *100))
    print('{} AUC :{}%'.format(name,auc*100))
    print('{} F1 :{}%'.format(name,f1*100))
    return accuracy,recall,f1
```

```python
from sklearn.metrics import mean_squared_error
def model_re(model,x_train,x_test,y_train,y_test,name):
    model = model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_pred,y_test)
    print('{} mse :{}%'.format(name,mse%100))
    return mse
```

```python
from sklearn.linear_model import LogisticRegression
#逻辑回归模型
model_fraud = LogisticRegression()
model_late = LogisticRegression()
model_stats(model_fraud,x_f_train,x_f_test,y_f_train,y_f_test,'Fraud')
model_stats(model_late,x_l_train,x_l_test,y_l_train,y_l_test,'late')
```
